```python
import bayesalpha as ba
import pandas as pd
import pyspark 
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[1]").getOrCreate()



data=spark.read.parquet("SPYpred.parquet")
data=data.toPandas()
display(data)
```

```python
data=data.set_index(data.Time).drop(['Close','Volume','Time','Ticker','features','indexedFeatures'], axis=1)
data.index=pd.to_datetime(data.index, unit='s')
display(data)
```

```python
max=data['prediction'].max()
print(max)
data['In']=data['prediction']/max
data=data.drop(data[data.In<=0].index)
data['Out']=data['In']*(1+data['label'])
data['Returns']=data['In']-data['Out']
data=data.drop(['label','prediction','In','Out'], axis=1)
display(data)#Returns is in percent from -1 to infinity, where 
#a 1% gain is 0.01
```

```python
trace=ba.fit_returns_population(data)

print(trace)
```

```python
trace.plot_prob()
print(trace.gains_pos_prob())
```
